```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('ExampleData.csv')
```

```python
df
```

```python
#Boş yerleri NaN olanları en iyi Linear Regression ile doldurabiliriz.
```

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
imp=IterativeImputer(estimator=lr)
imputed_data=imp.fit_transform(df)
```

```python
imputed_data
```

```python
from sklearn.impute import KNNImputer #En Yakın veri ile doldurma 
```

```python
knn=KNNImputer()
```

```python
imputed_data2=knn.fit_transform(df)
```

```python
pd.DataFrame(imputed_data2,columns=[df.columns])
```

```python
from sklearn.impute import SimpleImputer #Ortalama ile doldurma 
```

```python
imp=SimpleImputer(strategy='mean')#median,mode v.s
```

```python
imputed_data3=imp.fit_transform(df)
```

```python
pd.DataFrame(imputed_data3,columns=[df.columns])
```
