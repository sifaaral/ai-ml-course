# Scalling - Normalization -PCA

```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('pima-indians-diabetes.csv')
```

```python
df.head()
```

```python
# Birimleri farklı hepsini aynı skalaya getirmemiz gerekiyor Z skor da deniyor
```

```python
from sklearn.preprocessing import normalize,scale
```

```python
x=df.drop('Outcome',axis=1)
```

```python
x.head()
```

```python
yeni_x=scale(x) # - ile sıfır arasına alıyor buda normalizeden farkı bu  data büyükse -2 olabiliyor 
pd.DataFrame(yeni_x)
```

```python
normalizeddata=normalize(x)
```

```python
pd.DataFrame(normalizeddata) #Veriyi 0 ile 1 arasında alıyor veriyi normalize ediyor  bütün sütunlara eşit muamele etmesi için kullanılır bir 50 birisi 1 olduğundan normalize ediyoruz.
```

# Boyut Küçültme PCA - Principle Component Analyses

```python
# Sütun çoksa veriyi küçük hallere bölüyor  Ör. 5300 sütunlu bir veriyi nasıl fit edeceğiz zorlanır 
```

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA(2)
```

```python
x.head()
```

```python
data2=pca.fit_transform(x)
```

```python
data2
```

```python
pd.DataFrame(data2) # Vektörlerin izdüşümünü ala ala iki sütuna dönüştürüyor bir sürü sütunla uğraşmaktansa iki sütunla işi yapıyor 
```

```python
pd.DataFrame(data2, columns=['PCA1','PCA2']) 
```

```python
pca.explained_variance_ratio_ #verinin ne kadarı kayboldu 
```

```python
sum(pca.explained_variance_ratio_ )#verinin ne kadarı kayboldu 
```
